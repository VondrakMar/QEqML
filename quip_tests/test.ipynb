```python
from ase.io import read,write
from quippy.descriptors import Descriptor
import numpy as np
import sys
np.set_printoptions(threshold=sys.maxsize)

mol = read("W3.xyz@:",format="extxyz")
```

```python
soap = Descriptor("soap cutoff=3.5 l_max=5 n_max=6 atom_sigma=1.0 normalize=T delta=1.0 n_species=2 species_Z={1 8} n_Z=2 Z={1 8}")
soapO = Descriptor("soap cutoff=3.5 l_max=5 n_max=6 atom_sigma=1.0 normalize=T delta=1.0 n_species=2 Z=8 species_Z={1 8}")
soapH = Descriptor("soap cutoff=3.5 l_max=5 n_max=6 atom_sigma=1.0 normalize=T delta=1.0 n_species=2 Z=1 species_Z={1 8}")
soapO2 = Descriptor("soap cutoff=3.5 l_max=5 n_max=6 atom_sigma=1.0 normalize=T delta=1.0 n_species=1 n_Z=2 Z={1 8} species_Z={1 }")
soapH2 = Descriptor("soap cutoff=3.5 l_max=5 n_max=6 atom_sigma=1.0 normalize=T delta=1.0 n_species=1 n_Z=2 Z={8 1} species_Z={8}")
```

```python
calc_soap = soap.calc(mol)
calc_soapH = soapH.calc(mol)
calc_soapO = soapO.calc(mol)
calc_soapH2 = soapH2.calc(mol)
calc_soapO2 = soapO2.calc(mol)
```

```python
print(len(calc_soap[0]["data"]))
print(len(calc_soapH[0]["data"]))
print(len(calc_soapO[0]["data"]))
print(len(calc_soapH2[0]["data"]))
print(len(calc_soapO2[0]["data"]))
```
